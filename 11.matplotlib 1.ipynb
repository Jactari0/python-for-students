```python
# Polynomial regression ile ekstra bilgi
```

```python
from sklearn.linear_model import LinearRegression
```

```python
#linear regression model
LR = LinearRegression()

x = df1.I.values.reshape(-1,1)
y = df1.V.values.reshape(-1,1)
```

```python
LR.fit(x,y)
```

```python
b0 = LR.intercept_
b1 = LR.coef_
print("b0:",b0)
print("b1:",b1)
```

```python
print("y ekseninin kesildiği nokta = {}".format(LR.intercept_[0]))
print("---------")
print("doğrunun eğimi = {}".format(LR.coef_[0][0]))
print("---------")
print("Regresyon sonucu = {:.16f} x + {:.16f}".format(LR.coef_[0][0],LR.intercept_[0]))
```

```python
print("doğrunun eğimi, burada R Direncine karşı gelir = {}".format(LR.coef_[0][0]))
```

```python
def V0(x):
    return LR.intercept_ + LR.coef_ * x

V0(np.array(df1.I)).reshape(-1,1)
```

```python
df1["V calculated"] = V0(np.array(df1.I)).reshape(-1,1)
df1
```

```python
plt.figure(figsize = (12,8))

plt.title("Ohm Yasası")
plt.errorbar(df1.I, df1.V, xerr = df1.DI, yerr = df1.DV, label = "veri", 
            fmt = "o" ,color = "red", ecolor = "darkblue", elinewidth = 3, capsize=8)
plt.plot(df1["I"], df1["V calculated"], color = "green", label = "veri")
plt.xlabel("Akım (Amper)")
plt.ylabel("Gerilim (Volt)")
plt.legend()
plt.show
```
